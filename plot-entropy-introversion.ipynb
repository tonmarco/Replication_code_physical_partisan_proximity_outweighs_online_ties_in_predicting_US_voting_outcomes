```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format='retina' 
```

```python
# code for the font from https://github.com/scentellegher/code_snippets/blob/d6e3a65b0c4e715d982466f77d94c4f8b0827b38/matplotlib_font/Matplotlib_custom_font.ipynb

from matplotlib import font_manager

font_dirs = ['font/']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)

for font_file in font_files:
    font_manager.fontManager.addfont(font_file)

# set font
plt.rcParams['font.family'] = 'Barlow'
```

```python
# import data

df = pd.read_csv('./data/county_data.csv',  dtype = {'fips' : 'str', 'rucc_grouped' : 'str', 'RUCC_2013' : 'str'})
```

```python
plt.scatter(df['entropy_sci'], df['entropy_coloc'], s=1, c='#AF5A5B')
plt.xlim(0,1)
plt.ylim(0,1)
plt.xticks(size=13)
plt.yticks(size=13)
plt.xlabel('$D(i)_{online}$', size=14)
plt.ylabel('$D(i)_{offline}$', size=14)

sns.despine()
```

```python
plt.scatter(df['entropy_coloc'], (df['extroversion_coloc']), s=1, c='#AF5A5B')

plt.xscale('log')
plt.yscale('log')


plt.xticks(size=13)
plt.yticks(size=13)
plt.xlabel('$D(i)_{offline}$', size=14)
plt.ylabel('$E(i)_{offline}$', size=14)

sns.despine()
```

```python
plt.scatter(df['entropy_sci'], (df['extroversion_sci']), s=1, c='#AF5A5B')

plt.xscale('log')
plt.yscale('log')

plt.xticks(size=13)
plt.yticks(size=13)
plt.xlabel('$D(i)_{online}$', size=14)
plt.ylabel('$E(i)_{online}$', size=14)

sns.despine()
```
